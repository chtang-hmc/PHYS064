```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jv 
from scipy.special import yv
```

```python
def check_bessel_1():
    fig, ax = plt.subplots()
    for n in range(100):
        x = np.linspace(0.01, 10, 101) 
        ax.plot(x, jv(n-1, x)+jv(n+1,x)) 
        ax.plot(x, 2*n*jv(n, x)/x, '.', alpha=0.5) 
        ax.set_xlabel("$x$")
    plt.show()

check_bessel_1()

def check_bessel_1_analytic():
    x = np.linspace(0.01, 10, 101)
    for n in range(100):
        difference = jv(n-1, x)+jv(n+1,x) - 2*n*jv(n, x)/x
        assert np.all(np.abs(difference) < 1e-10)
        
check_bessel_1_analytic()
```

I used two methods to check this: first, by graphing it. All the points fit the graph really well. I also checked that the difference between two expressions are less than $1^{-10}$. No errors came from the assertion, so the errors are all within $1^{-10}$.

```python
def check_bessel_2():
    fig, ax = plt.subplots()
    for n in range(100):
        x = np.linspace(0.01, 50, 200) 
        ax.plot(x, jv(n-1, x)-jv(n+1,x)) 
        ax.plot(x, 2*differentiate_bessel(n,x), '.', alpha=0.5) 
        ax.set_xlabel("$x$")
    plt.show()
    
def differentiate_bessel(n, x):
    h = 1e-5
    return (jv(n, x+h) - jv(n, x-h))/(2*h)

check_bessel_2()

def check_bessel_2_analytic():
    x = np.linspace(0.01, 10, 101)
    for n in range(100):
        difference = jv(n-1, x)-jv(n+1,x) - 2*differentiate_bessel(n,x)
        assert np.all(np.abs(difference) < 1e-10)
        
check_bessel_2_analytic()
```

I used two methods to check this: first, by graphing it. All the points fit the graph really well. I also checked that the difference between two expressions are less than $1^{-10}$. No errors came from the assertion, so the errors are all within $1^{-10}$.
